```python
photo_extensions = ['.jpeg', '.jpg', '.png', '.webp']
photo_folder = tv_folder / 'Photos'

repeated_photos = []
if not photo_folder.exists():
    print(f'The folder {photo_folder.name} was created')
    photo_folder.mkdir(parents=True, exist_ok=True)
for folder in models.iterdir():
    destination_folder_photos = photo_folder / folder.name
    
    for photo in folder.rglob('*'):
        if photo.suffix in photo_extensions:
            repeated = False
            if not destination_folder_photos.exists():
                destination_folder_photos.mkdir(parents=True, exist_ok=True)
                print(f'The folder {destination_folder_photos.name} was created')
            destination_file = destination_folder_photos / photo.name
            counter = 1
            while destination_file.exists():
                destination_file = destination_folder_photos / f'{photo.stem} {counter:02d}{photo.suffix}'
                counter += 1
                repeated = True
            shutil.move(str(photo) , str(destination_file))
            if repeated:
                repeated_photos.append(photo)
if repeated_photos:
    print('Repeated photos')
    for ph in repeated_photos:
        print(ph)
```

```python
pyperclip.copy('')
exception_files = defaultdict(str)
destination_folder = tv_folder / 'Automatically Add To TV.localized'
continue_moving = True
error_folder = tv_folder / "Error_files"
error_folder.mkdir(parents=True, exist_ok=True)
skip_folder = []
start_process = time.time()
exception_files = defaultdict()

for folder in models.iterdir():
    if folder.is_dir() and folder.stem not in skip_folder:
        
        mp4_files_folder = [video for video in folder.rglob('*') if video.suffix.lower() in extensions]
        num_files = len(mp4_files_folder)

        for i, video_file in enumerate(mp4_files_folder,1):
            try:
            
                clear_output(wait=False)
                print(f'Folder {folder.name} ')
                message = f'Working on {i} / {num_files} {video_file.stem}'
                print(message)
                
                len_message = len(message)
                metadata = MP4(video_file)
                for key in list(metadata.keys()):
                    if key == '\xa9cmt' and not any(comment in metadata['\xa9cmt'][0] for comment in list_comments):
                        continue
                    del metadata[key]
                metadata['\xa9nam'] = clean_metadata(video_file.stem, list_substrings, dict_formated)
                if folder.stem != 'Delete Metadata':
                    metadata['\xa9ART'] = folder.name.replace('.', ' ')
                metadata.save()
    
    
                destination_file = destination_folder / video_file.name
                shutil.move(str(video_file) , str(destination_file))
                start_moving = time.time()
                while destination_file.exists():
                    current_time = time.time() - start_moving
                    if  current_time > 60 or pyperclip.paste()== 'stop_loop':
                        print(f'Breaking {video_file.stem}')
                        continue_moving = False
                        break
                        
                    print(duration_string(current_time), end='\r')
                    time.sleep(0.5)
            except Exception as e:
                error_destination = error_folder / video_file.name
                enumerator = 1
                while error_destination.exists():
                    error_destination = error_folder / f'{ideo_file.stem} {enumerator:02d}{ideo_file.suffix}'
                    enumerator += 1
                shutil.move(str(video_file) , str(error_destination))
                exception_files[video_file.stem] = e
                
        if not continue_moving:
            break
            
        clear_output(wait=False)
    if not continue_moving:
        break

deleted_files = []
duration_process = time.time() - start_process
print(f'The duration of the whole process -> {duration_string(duration_process)}')
extensions_delete = ['.nfo']

for file_path in tv_folder.rglob('*'):
    if file_path.suffix in extensions_delete or file_path.stem.startswith('.ghost'):
        
        try:
            file_path.unlink()
            deleted_files.append(file_path)
        except OSError as e:
            print(f"Error deleting file '{file_path}': {e}")
if deleted_files:
    print('The following files were deleted')
    for file in deleted_files:
        print(file)

do_not_delete_folder = [models / folder for folder in ['Delete Metadata', ]]

for subfolder in sorted(models.rglob("*"), reverse=True):
    
    if subfolder.is_dir() and not any(subfolder.iterdir()) and subfolder not in do_not_delete_folder:
        subfolder.rmdir()
        print(f"Deleted empty folder: {subfolder}")
if exception_files:
    print('Files with Error: ')
    for key, value in exception_files.items():
        print(f'{key}: {value}')
```
