# **Get the count of each Handle from a Playlist**

```python
searched_url = defaultdict(str)
source_playlist = choose_option(list_playlist_name, message = "Choose a Playlist: ")
source_id = next((d["id"] for d in playlist_names if source_playlist in d.values()), "")
print(f'{source_playlist}: {playlist_url}{source_id}')
start = time.time()
video_ids_playlist = get_all_ids_playlist(youtube, quota_filename, source_id, 20)
len(video_ids_playlist)
handles_dur_count = defaultdict(lambda: {"Duration": 0.0, 'Video IDs': []})
for video_id in video_ids_playlist[0:]:
    response = get_response_video_id(youtube, quota_filename, video_id)
    items = response.get('items',[])
    if not items:
        continue
    channelId = response['items'][0]['snippet']['channelId']
    snippet = items[0].get('snippet', {})
    publishedAt = snippet.get('publishedAt', "2005-04-24T03:31:52Z") #Timestamp of the first YouTube video ever published 
    contentDetails = items[0].get('contentDetails', {})
    duration_iso = contentDetails.get('duration', 'PT0S')
    duration = isodate.parse_duration(duration_iso).total_seconds()
    if channelId not in YT_content_creators['Channel ID'].values:
        if channelId not in searched_url:
            channel_response = get_channel_response(youtube, quota_filename, channelId)
            snippet_channel = channel_response['items'][0].get('snippet', {})
            handle = snippet_channel.get('customUrl', 'Error').replace('@', '')
            searched_url[channelId] = handle           
        else:
            handle = searched_url[channelId]
    else:
        handle = YT_content_creators[YT_content_creators['Channel ID'] == channelId]['Handle'].iloc[0]
    handles_dur_count[handle]['Duration'] += duration
    handles_dur_count[handle]['Video IDs'].append((video_id, publishedAt))
duration = time.time() - start
print(f'Duration => {duration_string(duration)}')
    
searched_url
```

```python
key_width = max(len(k) for k in handles_dur_count)
int_width = max(len(str(len(v["Video IDs"]))) for v in handles_dur_count.values())
float_width = max(len(f'{v["Duration"]}') for v in handles_dur_count.values())


# Print rows sorted by float descending
sorted_options = ['By Duration', "By Number of Videos", "By Alphabetical Order"]
sorted_option = choose_option(sorted_options, message = "Sort the Habdkes : ")

if sorted_option == sorted_options[0]:
    handles_sorted = sorted(handles_dur_count.items(), key=lambda x: x[1]["Duration"], reverse=True)
if sorted_option == sorted_options[1]:
    handles_sorted = sorted(handles_dur_count.items(), key=lambda x: len(x[1]["Video IDs"]), reverse=True)
if sorted_option == sorted_options[2]:
    asc_des = choose_option([True, False], message = "Reverse: ")
    handles_sorted = sorted(handles_dur_count.items(), key=lambda x: x, reverse=False)

second_col = 'No. Videos'
print(f'\033[1m {"Handle":<{key_width- 1}} | {second_col:>{int_width + len(second_col)}} | {"Duration":>{float_width}} | URL\033[0m')
total_duration = 0


for handle, v in handles_sorted:
    duration = v["Duration"]
    total_duration += duration
    channel_id = next((key for key, value in searched_url.items() if value == handle), None)
    if channel_id:
        url = 'https://www.youtube.com/channel/' + channel_id
    elif handle in YT_content_creators['Handle'].values:
        url = 'https://www.youtube.com/channel/' + YT_content_creators[YT_content_creators['Handle'] == handle]['Channel ID'].iloc[0]
    else:
        # This might not be necessay
        url = 'Not in Data Frame nor in the Searched dictionary'
    print(f'{handle:<{key_width}} | {len(v["Video IDs"]):>{int_width + len(second_col)}} | {duration_string(duration):>{float_width}} | {url}')
print(' \n')
print(f'The Total duration of the Playlist is {duration_string(total_duration)} ')
```

# **This would require at least 101 quotas per video**

```python
quota_limit = 8000
current_quota = get_today_quota(quota_filename, False)
min_quota_usage = sum(len(handles_dur_count[creator]['Video IDs']) for creator in handles_to_remove_add) * 101
if choose_option([True, False], f"Are you sure you want to proceed. Its gonna take at least {min_quota_usage:,} quotas and the current quota is {current_quota:,}"):
    initial_quota = get_today_quota(quota_filename, True)
    first_break = True
    for index, handle in enumerate(handles_to_remove_add, 1):
        video_ids = handles_dur_count[handle]['Video IDs']
        video_ids.sort(key=lambda x: x[1], reverse=False)
        if first_break:
            print(f'There are {len(video_ids)} from the handle {handle}')    
        for index_id, video_id_to_move in enumerate(video_ids, 1):
            message = f'{index:02d} / {len(handles_to_remove_add):02d} | {index_id:02d} / {len(video_ids):02d}'
            print(message, end='\r')
            if get_today_quota(quota_filename, False) > quota_limit:
                if first_break:
                    clear_output(wait=False) 
                    print(f'The process was breaked. There were {index - 1:,} videos moved. The follwing ID were not moved:')
                    first_break = False
                print(' '*len(message), end='\r')
                print(f'{handle}: ')
                for missing_index, video_id in enumerate(video_ids[index_id - 1:], 1):
                    print(f'\t{missing_index:02d} {yt_url}{video_id[0]}')
                break
            delelte_video_id_from_playlist(youtube, quota_filename, video_id_to_move[0], source_id, print_message = False)
            # add_video_to_playlist(youtube, source_id, video_id)
            if video_id_to_move[0] in videos_in_playlist:
                print(f'Video id already in playlist {index_id:02d} {yt_url}{video_id_to_move[0]} ')
                continue
            add_video_to_playlist(youtube, quota_filename, destination_id, video_id_to_move[0])
    final_quota = get_today_quota(quota_filename, True)
    print(f'There was {(final_quota - initial_quota):,} quoatas consumed')
else:
    clear_output(wait=False) 
    print('There was no video addded')
```
