```python
from pathlib import Path
from mutagen.mp4 import MP4
import shutil
import time
from IPython.display import clear_output
from writing import *


download_folder = Path('/Volumes/ROBERTO/Downloads')
if not download_folder.exists():
    download_folder.mkdir(parents=True, exist_ok=True)
    print(f'Folder {download_folder} was created')

of_folder = Path('OF')
if not of_folder.exists():
    of_folder.mkdir(parents=True, exist_ok=True)
    print(f'Folder {of_folder} was created')

def get_integer_input(prompt="Number of files: "):
    while True:
        user_input = input(prompt)
        try:
            return int(user_input)
        except ValueError:
            clear_output()
            print("Invalid input. Please enter an integer.", end='\r')

start_process = time.time()
```

```python
content_creator = input('Content Creator: ').strip()
```

```python
usename = input('Username:').strip().replace('@','').replace('https://onlyfans.com/', '')
file_path = of_folder / usename
used_metadata = get_elements_from_file(file_path)
used_metadata
```

```python
date_string = input('Date string: ').strip()
date_number = write_date_numbers(date_string)
date_number
```

```python
num_files = get_integer_input(prompt="Number of videos in the Bundle: ")



if num_files == 1:
    metadata_name = f'{date_number}{usename}'
    add_element_to_file(file_path, f'{date_number}{usename} 01', sort_list = True, print_statement = False)
    files_to_download = [download_folder / f'{date_string}.mp4']
else:
    metadata_name = f'{date_number}{usename} 01'
    add_element_to_file(file_path, f'{date_number}{usename}', sort_list = True, print_statement = False)
    start_enumerating = get_integer_input(prompt="start_ennumerationg: ")
    files_to_download = [download_folder / f'{date_string} {i:02d}.mp4'   for i in range(start_enumerating, num_files + 1)]
files_to_download
```

```python
comment = input('Comment: ').strip()
```

```python
pyperclip.copy('')
destination_folder = tv_folder / 'Automatically Add To TV.localized'
stop_process = False
for video in files_to_download:
    # Wait for the file to download 
    pyperclip.copy(video.stem)
    start_waiting = time.time()
    print(f'Waiting for {video.stem}')
    while not video.exists():
        current_time = time.time() - start_waiting
        if  current_time > 120 or pyperclip.paste()== 'stop_loop':
            print(f'Not found {video.name}')
            stop_process = True
            break
        time.sleep(1)
        print(duration_string(time.time()-start_waiting), end='\r')
    if stop_process:
        print('StoToday 01ped in the waiting')
        break

    #Write the metadata
    print(f'Writing metadata for {video.stem}')
    used_metadata = get_elements_from_file(file_path)
    metadata = MP4(video)
    metadata.clear()
    metadata['\xa9ART'] = content_creator
    counter = 1
    while metadata_name in used_metadata:
        
        metadata_name = f'{date_number}{usename} {counter:02d}'
        counter += 1
        if pyperclip.paste()== 'stop_loop':
            stop_process = True
            break
        
    if stop_process:
        print('Stoped in the Metadata')
        break
    metadata['\xa9nam'] = metadata_name
    metadata['\xa9cmt'] = comment
    metadata.save()
    add_element_to_file(file_path, metadata_name, sort_list = True, print_statement = False)

    # Move file
    print(f'Moving {video.stem}')
    destination_file = destination_folder / video.name

        
    shutil.move(str(video) , str(destination_file))
    start_moving = time.time()
    while destination_file.exists():
        current_time = time.time() - start_moving
        if  current_time > 60 or pyperclip.paste()== 'stop_loop':
            print(f'Breaking {video_file.stem}')
            stop_process = True
            break
            
        print(duration_string(current_time), end='\r')
        time.sleep(0.5)
    if stop_process:
        print('Stoped in the move')
        break
    clear_output(wait=False)
    
```
